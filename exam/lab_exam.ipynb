```python
import pandas as pd
```

```python
diamond_data = pd.read_csv("Diamond_price.csv")
```

```python
diamond_data.head()
```

```python
# Find null values
null_values = diamond_data.isnull().sum()
print(null_values)
# null_values.describe()

# Fill null values with mean
numeric_cols = diamond_data.select_dtypes(include=['float64', 'int64']).columns
means = diamond_data[numeric_cols].mean()
diamond_data[numeric_cols] = diamond_data[numeric_cols].fillna(means)
print("\nNull Values After Filling:")
print(diamond_data.isnull().sum())
```

### determining the count of each diamond cut

```python
cut_counts = diamond_data['Cut(Quality)'].value_counts()
cut_counts
```

### analyzing the frequency of different diamond colors

```python
color_counts = diamond_data['Color'].value_counts()
color_counts
```

### counting the occurrences of each clarity grade

```python
clarity_counts = diamond_data['Clarity'].value_counts()
clarity_counts
```

### To show records where the cut quality is 'Ideal' and the diamond color is 'D'

```python
ideal_diamonds = diamond_data[(diamond_data['Cut(Quality)'] == 'Ideal') & (diamond_data['Color'] == 'D')]

ideal_diamonds
```

### To remove records where the carat weight of the diamond is less than 0.5:

```python
diamond_data = diamond_data[diamond_data['Carat(Weight of Daimond)'] >= 0.5]
diamond_data
```

### To calculate the mean, median, and standard deviation of diamond prices:

```python
mean_price = diamond_data['Price(in US dollars)'].mean()
median_price = diamond_data['Price(in US dollars)'].median()
std_price = diamond_data['Price(in US dollars)'].std()
```

### To determine the variance of diamond prices based on different clarity grades

```python
variance_by_clarity = diamond_data.groupby('Clarity')['Price(in US dollars)'].var()
variance_by_clarity
```

### To create a new feature named 'Volume' by multiplying the length, width, and depth of the diamonds

```python
diamond_data['Volume'] = diamond_data['X(length)'] * diamond_data['Y(width)'] * diamond_data['Z(Depth)']
diamond_data['Volume']
```
